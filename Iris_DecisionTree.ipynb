<H3 style="color:Crimson;"> Loading Required Python Libraries </H3>

```python
import pandas as pd
import numpy as np
import sklearn.datasets as datasets
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
```

<h3 style="color:Crimson;">loading and checking the dataset</h3>

```python
data=datasets.load_iris()
data.keys()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df
```

```python
df.info()
```

<b> So there are no null values. </b>

<h3 style="color:Crimson;">Splitting the Dataset into Train and Test</h3>

```python
X = df
y = data.target
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=0,test_size=.2)
```

<h3 style="color:Crimson;"> Fitting the Decision Tree</h3>

```python
dtree=DecisionTreeClassifier(max_depth=3,criterion='entropy')
dtree.fit(X_train,y_train)
y_predict = dtree.predict(X_test)

print("Fitted the Decision Tree Classifer!")
```

<h3 style="color:Crimson;"> Model Evaluation </h3>

<li><b> Confusion Matrix </b></li>

```python
confusion_matrix(y_test,y_predict) 
```

<li><b> Classification Report </b></li>

```python
print(classification_report(y_test,y_predict))
```

<li><b> Accuracy Score </b></li>

```python
print("Accuracy Score:%.2f"%accuracy_score(y_test,y_predict))
```

<b>We see that the Precision, Recall, F1-Score are above 80% also we have 97% accuracy score which means the model is fitted well and works nicely.</b>

<h3 style="color:Crimson;"> Visualizing the Decision Tree</h3>

```python
plt.figure(figsize=(18,12))
tree.plot_tree(dtree,feature_names=data.feature_names, class_names = data.target_names, filled=True,rounded=True)
plt.title("Decision Tree",fontsize=25,color="GRAY")
plt.show()
```

<h3 style="color:Crimson;"> User Input Prediction</h3>

```python
input0 = input("Please input Sepal Length, Sepal Width, Petal Length, Petal Width in centemeters, separated by commas:").split(',' or ', ' or ' ,')
input1=[]
for i in input0:
    input1.append(float(i))
print(input1)    
Predicted_Class = dtree.predict([input1])
print(data.target_names[Predicted_Class][0])
```
